```python
#2. Desarrolla un programa que encuentre y muestre todos los números perfectos en el
#rango del 1 al 1,000,000. Un número perfecto es un entero positivo que es igual a la
#suma de sus divisores propios positivos. Los divisores propios son todos los
#divisores del número, excluyendo al propio número.
#El programa debe imprimir cada número perfecto encontrado.

import os

os.system("cls")

def encontrar_numeros_perfectos(limite):

    print(f"Buscando números perfectos en el rango de 1 a {limite}...")


    for numero in range(2, limite + 1):
        suma_divisores = 1


        for i in range(2, int(numero*0.5) + 1):
            if numero % i == 0:

                suma_divisores += i

                if ii != numero:
                    suma_divisores += numero // i


        if suma_divisores == numero:
            print(f"Número perfecto encontrado: {numero}")


LIMITE_SUPERIOR = 1000000


encontrar_numeros_perfectos(LIMITE_SUPERIOR)
```

```python
#3. En un curso básico de Geometría Euclidiana se quiere construir un software que
#facilite el cálculo de áreas de ciertas figuras geométricas y algunos otros elementos
#importantes de la geometría básica. Para esto, se le pide que construya un software
#que pueda acceder a esas diferentes opciones a través de un menú. Que en cada
#submenú se pida al usuario que introduzca lo que se necesita para hacer los cálculos
#correspondientes.
#Este software será utilizado por personas con poca experiencia en informática, por
#lo que se le pide que sea fácil de maneja y que los mensajes sean muy intuitivos.

#** MENU **
#Área de un triangulo
#Área de un círculo
#N-ésimo término de Fibonacci
#Los 100 primeros números primos
#Cálculo del número PI
#Salir
#>> Ingrese una opción?

import os
isActive = True


def menu():
    print("----------------------------------------------------")
    print("BIENVENIDO AL CURSO DE GEOMETRÍA EUCLIDIANA")
    print("** MENU **")
    print("1. Área de un triangulo")
    print("2. Área de un círculo")
    print("3. N-ésimo término de Fibonacci")
    print("4. Verificar si un número es primo")
    print("5. Los 100 primeros números primos")
    print("6. Cálculo del número PI")
    print("0. Salir")
    print("----------------------------------------------------")
    print("")
    

def areaTriangulo(b, h):
    os.system("clear")
    b = float(input("Ingrese la base del triangulo: "))
    h = float(input("Ingrese la altura del triangulo: "))
    area = (b*h)/2
    print(f"El área del triangulo es {area}")
    input("\nPresiona Enter para continuar...")
    os.system("clear")
    return area   
    
def areaCirculo(r):
    os.system("clear")
    r = float(input("Ingrese el radio del círculo: "))
    area = 3.1416 * (r**2)
    print(f"El área del círculo es {area}")
    input("\nPresiona Enter para continuar...")
    os.system("clear")
    return area
    

def fibonacci():
    os.system("clear")
    a = 0
    b = 1
    cantidad = 10
    print("La serie de Fibonacci es: ", end="")
    for i in range(cantidad):
        if i == 0:
            print(a, end="")
        elif i == 1:
            print(f", {b}", end="") 
        else:
            siguiente = a + b
            print(f", {siguiente}", end="")
            a = b
            b = siguiente
    input("\nPresiona Enter para continuar...")
    os.system("clear")

def es_primo(num):

    if num < 2:
        return False
    for i in range(2, num + 1):
        if num % i == 0:
            return False
    return True

def calcularPi(interaciones):
    interaciones=1_000_000
    pi=0
    for i in range(interaciones):
        pi += ((-1)** i) / (2 *i + 1)

    pi*= 4
    print(f"El valor aproximado de pi con 8 decimales es : {pi:.8f}")

while isActive:
    menu()
    opcion = int(input("Ingrese una opción: "))
    match opcion:
        case 1:
            areaTriangulo(0, 0)
        case 2:
            areaCirculo(0)
        case 3:
            fibonacci()
        case 4:
            Num = int(input("Ingrese un número para verificar si es primo: "))
            if es_primo(Num):
                print(f"El número {Num} es primo.")
            else:
                print(f"El número {Num} no es primo.")
            input("\nPresiona Enter para continuar...")
            os.system("clear")
        case 5:
            pass
        case 6:
            calcularPi(1_000_000)
        case 0:
            isActive = False
            print("Gracias por usar el programa. Hasta luego!")
        case _:
            print("Opción no válida, intente de nuevo.")

```
